```python
#it looks like the county code is in the county_code column. it is our goal to create FIPS codes out of these county codes.
    #FIPS codes uniquely identify counties by appending a 3-digit county code to a 2-digit state code.
    #we are only working with California data, so we can always use "06" for the state code.
    #problem: we need to edit the county codes to all be 3 digits, with 0s filling in the first entries if necessary
    #let's write a function for that
    
def fips_column(df, county_column_name):
    j = 0
    df["fips"] = df[county_column_name]
    for i in df[county_column_name]:
        if len(f"{i}") == 1:
            df["fips"][j] = "06" + "00" + f"{i}"
        elif len(f"{i}") == 2:
            df["fips"][j] = "06" + "0" + f"{i}"
        else:
            df["fips"][j] = "06" + f"{i}"
        j += 1
```

```python
fips_column(PM2_2009_2014, "county_code")
fips_column(PM10_2009_2014, "county_code")
fips_column(PM2_2014_2021, "county_code")
fips_column(PM10_2014_2021, "county_code")
```
